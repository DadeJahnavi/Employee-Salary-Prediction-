```python
x = df[["Years_at_Company", "Satisfaction_Level", "Average_Monthly_Hours"]]
y=df["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
x_test
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error

def results(predictions):
  print("Mean abousulete error on model is {}", format(mean_absolute_error(y_test,predictions)))
  print("Root mean squared error on the model is {}", format(np.sqrt(mean_squared_error(y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
predictionlr = lr.predict(x_test)
```

```python
results(predictionlr)
```

### **Support Vector Regression**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
svrmodel = SVR()
```

```python
param_gridsvr={
    'C': [0.1, 1, 10],
    'epsilon': [0.1, 0.2, 0.5],
    'kernel':['linear', 'poly', 'rbf'],
    'degree': [2, 3, 4]
}
```

```python
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
```

```python
gridsvr.fit(x_train, y_train.values.ravel())
```

```python
gridsvr.best_params_
```

```python
predictionssvr = gridsvr.predict(x_test)
```

```python
results(predictionssvr)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rfmodel = RandomForestRegressor()
```

```python
param_gridrfr={
    "n_estimators": [2,3,4,5,6],
    "max_depth":[5,10,15]
}
```

```python
gridrfr= GridSearchCV(estimator=rfmodel, param_grid=param_gridrfr, cv=5)
```

```python
gridrfr.fit(x_train, y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
predictionofrfr = gridrfr.predict(x_test)
```

```python
results(predictionofrfr)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have MAE and RMSE values for each model
mae_lr = mean_absolute_error(y_test,predictionlr)
rmse_lr = np.sqrt(mean_squared_error(y_test,predictionlr))

mae_svr = mean_absolute_error(y_test,predictionssvr)
rmse_svr = np.sqrt(mean_squared_error(y_test,predictionssvr))

mae_rfr = mean_absolute_error(y_test,predictionofrfr)
rmse_rfr = np.sqrt(mean_squared_error(y_test,predictionofrfr))


labels = ['Linear Regression', 'Support Vector Regression', 'Random Forest Regressor']
mae_values = [mae_lr, mae_svr, mae_rfr]
rmse_values = [rmse_lr, rmse_svr, rmse_rfr]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, mae_values, width, label='MAE')
rects2 = ax.bar(x + width/2, rmse_values, width, label='RMSE')

ax.set_ylabel('Error Value')
ax.set_title('Model Comparison: MAE and RMSE')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.tight_layout()
plt.show()
```
